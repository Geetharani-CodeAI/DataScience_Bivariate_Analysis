```python
df = pd.read_csv("Zomato_Chennai.csv")
df
```

# 2) Data Cleaning

#### a)Dropping the redundant columns:

```python
# Dropping Unnecessary URl column for analysis

df.drop(['Zomato URL'], axis = 1, inplace = True)
```

```python
df
```

```python
df.rename(columns = {'Price for 2':'Price'},inplace = True)
```

#### b) Handling Null values

```python
df.isnull().sum()
```

```python
# Replacing nulls with 'NA'

df.dropna(inplace = True)
df.isnull().sum()
```

#### c) Converting any non-numerical values into Numerical datatypes

```python
df['Dining Rating'].value_counts()
```

```python
df['Delivery Rating'].value_counts()
```

```python
df['Dining Rating Count'].value_counts()
```

```python
df['Delivery Rating Count'].value_counts()
```

```python
df.info()
```

```python
df['Dining Rating'] = df['Dining Rating'].astype('float64')
df['Dining Rating Count'] = df['Dining Rating'].astype('int64')
df['Delivery Rating'] = df['Dining Rating'].astype('float64')
df['Delivery Rating Count'] = df['Dining Rating'].astype('int64')
```

#### d) Working with Location Column

```python
df['Location'].value_counts()
```

```python
# splitting the location column and storing it in a temp df

temp_df = df['Location'].str.split(',', n =1, expand = True)
temp_df
```

```python
df['Location'] = temp_df[0]
df['Location']
```

#### e) Removing duplicates

```python
df[df.duplicated]
```

```python
df = df.drop_duplicates()
```

```python
df
```

#### f) Removing Dining Rating or Delivery Rating is 0

```python
useless_data = (df['Dining Rating'] == 0)|(df['Delivery Rating'] == 0)
df = df[~useless_data]
```

```python
# Removing Invalid columns in all columns

df = df[df.ne('Invalid').all(axis=1)]
df
```

# 5. Performing Exploratory Data Analysis

#### 1. Which are the Top 10 highest rated Biryani Restaurant in Chennai?

```python
biryani_df = df[df['Cuisine'].str.contains('Biryani')]
biryani_df.head()
```

#### 2. How many restaurant have Mutton Biryani?

```python
Mbiryani_df = df[df['Top Dishes'].str.contains('Mutton Biryani')]
Mbiryani_df
```

#### 3. What are the percentage of restaurants  by Dining Rating in Chennai?

```python
dining_rating_df = df['Dining Rating'].value_counts().reset_index()
dining_rating_df.rename(columns ={'count':'Count of Restaurants'})
```

```python
sns.jointplot(dining_rating_df, x = 'Dining Rating', y = 'count', kind = 'hex')
plt.show()
```

#### 4. Restaurants have Car Parking features in Chennai?


```python
park_df = df[df['Features'].str.contains('Free Parking')]
park_df.sort_values(by = 'Dining Rating',ascending = False)
```

#### 5. Which Restaurant has high dining rated and low price?

```python
high_dining_df = df[df['Dining Rating'] >= 4.5]
high_dining_df.head()
```

```python
print((df['Price'].max())/5)
```

```python
# Low Price means below mean value

low_price_df = df[['Name of Restaurant','Location','Cuisine','Top Dishes','Price','Dining Rating','Dining Rating Count']]
low_price_df = low_price_df[low_price_df['Price'] <= 600 ]
low_price_df
```

```python
# MERGING high_dining_rating and less price df

high_ratelow_df = pd.merge(high_dining_df, low_price_df, how = 'inner', on = ['Name of Restaurant','Location'])
high_ratelow_df = high_ratelow_df[['Name of Restaurant','Price_x','Cuisine_x','Location','Top Dishes_x','Dining Rating_x','Dining Rating Count_x']] # suffix appears automatically while doing merge
high_ratelow_df.rename(columns = {'Name of Restaurant':'Name of Restaurant','Price_x':'Price','Cuisine_x':'Cuisine',
                                   'Top Dishes_x':'Top Dishes','Dining Rating_x':'Dining Rating','Dining Rating Count_x':'Dining Rating Count'},inplace = True)
```

```python
# High rated Low Price Affordable Restaurant
high_ratelow_df
```

```python
fig = px.histogram(high_dining_df,x = 'Location', color = 'Cuisine',
                  title = 'No. of Best Restaurant for each Cuisine Type by Places').update_xaxes(categoryorder='total descending')
fig.show()
```

#### 6. How many High Rated Vegetarian Only Restaurants in Chennai?


```python
veg_df = df[df['Features'].str.contains('Vegetarian Only')]
veg_df
```

```python
veg_dining_df = veg_df[veg_df['Dining Rating'] >= 4.5]
veg_dining_df
```

```python
fig = px.scatter(veg_dining_df, x = 'Location', y = 'Price', color = 'Cuisine', symbol='Cuisine',
                title = "Avg Price Distribution of High rated Vegetarian Restaurant for each Cuisine Type").update_traces(marker_size =10)
fig.show()
```

#### 7. How many Restaurant has Mexican Style Cuisine?

```python
mexi_df = df[df['Cuisine'].str.contains('Mexican')]
mexi_df
```

```python
fig = px.bar(mexi_df, x = 'Location', y = 'Delivery Rating', color = 'Price', title = 'Mexican Restaurants in Chennai')
fig.show()
```

#### 8. Is there a relation between Price and Dining Rating by each Cuisine Type?

```python
price_dining_df = df.groupby(['Cuisine','Dining Rating'])['Price'].mean().reset_index()
price_dining_df
```

```python
fig = px.scatter(price_dining_df, x = 'Dining Rating', y = 'Price', color = 'Cuisine')
fig.show()
```

#### 9. Restaurants with Valet Parking

```python
valet_df = df[df['Features'].str.contains('Valet Parking')]
valet_df.groupby(by = 'Location').agg({'Name of Restaurant':'count','Price':'mean'}).reset_index().rename(columns = {'Name of Restaurant':'Count'})
```

```python
fig = px.histogram(valet_df, x ='Location', color = 'Price', 
        title = 'No. of Restaurants by Valet Parking', labels = {'Count':'Count'})
fig.show()
```

#### Q10) What are the percentage of Restaurants by Rating Type in Chennai?

```python
dining_rating_df = df['Dining Rating'].value_counts().reset_index()
dining_rating_df.rename(columns = {'index':'Dining Rating', 'count': 'Count of Restaurant'},inplace = True)
dining_rating_df
```

```python
fig = px.pie(dining_rating_df, names = 'Dining Rating', values = 'Count of Restaurant', color = 'Dining Rating', 
            title = 'Percentage of Restaurants by Dining Rating').update_traces(textposition='inside', textinfo = 'percent+label')
fig.show()
```
