# Data Cleaning

#### a) Removing redundant information from the dataset

```python
# Checking redundant rows
# this code is pulling out rows where the column accidentally contains its own header name "PAGE NO" instead of actual page numbers
wrong_data = raw_df['PAGE NO'] == 'PAGE NO'
raw_df[wrong_data]
```

```python
# Performing negation of the wrong dataset so it removes wrong data and keeping the correct data alone
raw_df = raw_df[~wrong_data]
raw_df
```

```python
# Dropping the not required columns for analysis
raw_df.drop(['URL','PAGE NO','CITY'],axis = 1,inplace = True)
raw_df
```

```python
raw_df.head()        # Returns first 5 rows
```

#### b) Removing Null records

```python
raw_df.isnull().sum()
```

```python
# Checking null row from price column
raw_df[raw_df['PRICE'].isnull()]
```

```python
raw_df = raw_df.drop(labels = 15080,axis = 0)
# raw_df[raw_df['PRICE'].isnull()]
```

```python
# Replacing other null values with NA
raw_df.fillna('NA',inplace = True)
```

```python
# Checking for null again
raw_df.isnull().sum()
```

#### c) Converting any non-numerical values into Numerical datatypes

```python
# Replacing the text values with '0'
raw_df['RATING'].value_counts()
```

```python
# New and Opening are text values so replacing with '0'
raw_df['RATING'].replace(to_replace=['-','NEW','Opening'], value = '0', inplace = True)
raw_df["RATING"].value_counts()
```

```python
# Check votes column for any New or Opening
raw_df['VOTES'].value_counts()
```

```python
raw_df['VOTES'].replace(['-','NEW','Opening'], value = 0,inplace= True)
#raw_df['VOTES'].value_counts()
```

```python
# Changing data type of numerical columns
raw_df['PRICE'] = raw_df['PRICE'].astype('int64')
raw_df['RATING'] = raw_df['RATING'].astype('float64')
raw_df['VOTES'] = raw_df['VOTES'].astype('int64')
```

```python
raw_df['RATING'].dtype
```

```python
raw_df.info()
```

#### d) Working with Timing column

```python
raw_df['TIMING'].value_counts()
```

```python
# Splitting the column and storing it in a temp_df
# Splits the string in each row wherever the first "(" occurs.
# n=1 → Only splits once (into at most 2 parts).

temp_df = raw_df['TIMING'].str.split("(",n=1,expand = True) 
temp_df
```

```python
# Assigning columns to the raw dataframe 
raw_df['TIMING'] = temp_df[0]
raw_df['DAYS_OPEN'] = temp_df[1]
raw_df.head()
```

```python
raw_df['DAYS_OPEN'] = raw_df['DAYS_OPEN'].str.replace(')','',regex = False)  # Regex - Regular Expression that uses a search pattern to find a string or set of string
raw_df.head()
```

```python
# Checking null values in DAYS_OPEN
raw_df.isnull().sum()
```

```python
# Replacing DAYS_OPEN null with 'NA'

raw_df.fillna('NA',inplace = True)
raw_df.isnull().sum()
```

```python
raw_df.info()
```

#### e) Removing Rating or Votes is 0

```python
# Finding those restaurant whose rating or votes is 0

useless_data = (raw_df['RATING'] == 0.0) | (raw_df['VOTES'] == 0)
raw_df[useless_data]
```

```python
# Performing Negation
raw_df = raw_df[~useless_data]
raw_df
```

#### f) Working On RATING_TYPE Column

```python
# Checking RATING_TYPE for null

raw_df['RATING_TYPE'].value_counts()
```

```python
# Translating the texts into proper English text

raw_df["RATING_TYPE"].replace(to_replace='Excelente',value = 'Excellent',inplace = True)
raw_df["RATING_TYPE"].replace(to_replace=['Veľmi dobré','Bardzo dobrze','Muy Bueno','Velmi dobré'],value = 'Very Good',inplace = True)
raw_df['RATING_TYPE'].replace(to_replace=['Skvělá volba','Dobrze','Bueno','Buono','Dobré','Bom','Skvělé'] , value='Good', inplace = True)
raw_df['RATING_TYPE'].replace(to_replace=['Priemer','Média','Çok iyi'] , value='Average', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Průměr','Promedio','Ortalama','Muito Bom','İyi'] , value='Poor', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Baik','Biasa','Media','Sangat Baik'] , value = 'Very Poor', inplace =True)
```

```python
raw_df['RATING_TYPE'].value_counts()
```

#### g) Working on 'REGION' Column

```python
raw_df['REGION'].value_counts()
```

```python
# Removing the irrelevant text from the column 'REGION'

raw_df['REGION'] = raw_df['REGION'].str.replace('[a-zA-Z].+--','',regex = True )
```

```python
raw_df['REGION'].value_counts()
```

```python
# Removing the West & East from the REGION Column

raw_df['REGION'] = raw_df['REGION'].str.replace('West| west| East| east','',regex = True)
```

```python
raw_df['REGION'].value_counts()
```

```python
# Replacing Small regions with Known region name

raw_df['REGION'] = raw_df['REGION'].str.replace('4 Bungalows|7 Andheri|Azad Nagar|Near Andheri Station|Veera Desai Area','Andheri',regex = True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Bandra Kurla Complex','Bandra',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('CBD-Belapur|Byculla|Churchgate|Bhayandar','CBD Belapur',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Girgaon Chowpatty','Chowpatty',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Dadar Shivaji Park','Dadar',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Flea Bazaar Café|Kamala Mills Compound','Lower Parel',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Runwal Green','Mulund',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Mumbai CST Area','Mumbai Central',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Kopar Khairane|Seawoods|Turbhe|Ulwe','Navi Mumbai',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('New Panvel|Old Panvel','Panvel',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Kamothe','Sion',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Ghodbunder Road|Majiwada','Thane',regex=True)
raw_df['REGION'] = raw_df['REGION'].str.replace('Kharghar','Khar',regex=True)
```

#### h. Removing Duplicate Records

```python
# Finding all the duplicate rows

raw_df[raw_df.duplicated()]
```

```python
# Dropping all the duplicate rows

raw_df = raw_df.drop_duplicates()
```

#### Q10) Find the list of Affordable Restaurants?

Criteria: 1) Low Price   2) High Rated

```python
max_price = df['PRICE'].max()
#max_price
one_fourth_price = max_price/4
one_fourth_price
```

```python
# Finding List of restaurant have 1/4 th or less than that

aff_rest_df = df[['NAME',"PRICE","CUSINE_CATEGORY","REGION","CUSINE TYPE"]]
aff_rest_df = aff_rest_df[aff_rest_df['PRICE'] <= 1250]
aff_rest_df
```

```python
# Finding High rated List of restaurants

high_rated_df = df[['NAME',"PRICE","CUSINE_CATEGORY","REGION","CUSINE TYPE",'RATING']]
high_rated_df = high_rated_df[high_rated_df['RATING']>=4.5]
high_rated_df.sort_values(by = "PRICE",inplace = True)
high_rated_df
```

```python
# Merging aff_rest_df and high_rated_df

high_rated_aff_df = pd.merge(aff_rest_df, high_rated_df, how = "inner", on = ['NAME', 'REGION'])
high_rated_aff_df = high_rated_aff_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x']] # suffix appears automatically while doing merge
high_rated_aff_df.rename(columns = {'NAME':'NAME','PRICE_x':'PRICE','CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
                                   'REGION':'REGION','CUSINE TYPE_x':'CUSINE TYPE'},inplace = True)
```

```python
# Affordable restaurant with low price and high rating
high_rated_aff_df
```

#### Q10) Find the list of most Reliable Restaurants?

Criteria: 1) Low Price 2)High Rated 3) large no.of votes

First Step: Find the restaurants with Votes greater than Mean of Votes


```python
mean_votes = df['VOTES'].mean()
print(mean_votes)
```

```python
# Finding List of restaurants votes greater than or equal to mean of votes

mean_rest_df = df[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','VOTES']]
mean_rest_df = mean_rest_df[mean_rest_df['VOTES'] > 177]
mean_rest_df.sort_values(by ='VOTES',inplace = True)
mean_rest_df
```

#### To Obtain Most Reliable Restaurant,

By Simply taking the intersection of high_rated_aff_df and mean_rest_df
This dataframe has 
1. Cost is below 1250
2. Rating is above 4.5
3. Votes are above 177

```python
reliable_rest_df = pd.merge(mean_rest_df, high_rated_aff_df, how= 'inner', on=['NAME','REGION'])
reliable_rest_df = reliable_rest_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x']]
reliable_rest_df.rename(columns = {'NAME':'NAME','PRICE_x':'PRICE','CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
                                   'REGION':'REGION','CUSINE TYPE_x':'CUSINE TYPE'},inplace = True)
```
